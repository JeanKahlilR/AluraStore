```python
df_loja1 =pd.DataFrame()
df_loja2 =pd.DataFrame()
df_loja3 =pd.DataFrame()
df_loja4 =pd.DataFrame()
```

```python
df_loja1['ano'] = df_loja1['Data_Compra'].dt.year
df_loja1['mes'] = df_loja1['Data_Compra'].dt.month
```

```python
df_loja2['ano'] = df_loja2['Data_Compra'].dt.year
df_loja2['mes'] = df_loja2['Data_Compra'].dt.month
```

```python
df_loja3['ano'] = df_loja3['Data_Compra'].dt.year
df_loja3['mes'] = df_loja3['Data_Compra'].dt.month
```

```python
df_loja4['ano'] = df_loja4['Data_Compra'].dt.year
df_loja4['mes'] = df_loja4['Data_Compra'].dt.month
```

```python
agreg_loja1 = df_loja1.groupby(['ano', 'mes']).sum('Preco').reset_index()
agreg_loja2 = df_loja2.groupby(['ano', 'mes']).sum('Preco').reset_index()
agreg_loja3 = df_loja3.groupby(['ano', 'mes']).sum('Preco').reset_index()
agreg_loja4 = df_loja4.groupby(['ano', 'mes']).sum('Preco').reset_index()
```

```python
agreg_loja1 = agreg_loja1.rename(columns={'Preco': 'Venda_Loja1'})
agreg_loja2 = agreg_loja2.rename(columns={'Preco': 'Venda_Loja2'})
agreg_loja3 = agreg_loja3.rename(columns={'Preco': 'Venda_Loja3'})
agreg_loja4 = agreg_loja4.rename(columns={'Preco': 'Venda_Loja4'})
```

```python
faturamento_geral = agreg_loja1.merge(agreg_loja2, on = ['ano', 'mes'], how = 'outer')
faturamento_geral = faturamento_geral.merge(agreg_loja3, on = ['ano', 'mes'], how = 'outer')
faturamento_geral = faturamento_geral.merge(agreg_loja4, on = ['ano', 'mes'], how = 'outer')
```

```python
def produto_mais_menos(loja, produto, qtd, maior, menor):
    print(f'PRODUTOS MAIS VENDIDOS {loja.upper()}:')
    for i,j in zip(produto,qtd):
        if j == max(qtd):
            print(f'Produto: {i} | Qtd: {j}')
            maior[i] = j
    print()
    print(f'PRODUTOS MENOS VENDIDOS {loja.upper()}:')
    for i,j in zip(produto,qtd):
        if j == min(qtd):
            print(f'Produto: {i} | Qtd: {j}')
            menor[i] = j  
               
```

# Relatorio: Recomendação de Venda de Loja

#### 1 - Introdução:

Com o objetivo de auxiliar o Senhor Jõao na decisão de qual loja vender foi feito a analise dos seguintes fatores:
- Faturamento das lojas
- Vendas por categoria
- Média de avaliação dos cliente
- Produtos mais e menos vendidos
- Frete médio por loja

#### 2 - Análise dos dados:


2.1 - Faturamento Total por Loja:

A Loja 4 possui o menor faturamento total. Além disso, ao analisar o histórico de vendas, observa-se que seu desempenho sempre esteve na margem inferior em comparação às demais lojas. Nos últimos cinco meses, houve ainda uma tendência de queda no volume de vendas. Caso o objeito seja vender a loja menos lucrativa essa seria a principal candidata.


2.2 - Vendas por Categoria:

Os dados não revelam discrepâncias significativas na distribuição de vendas por categoria entre as lojas. Portanto, esse fator não influencia decisivamente na escolha da loja a ser vendida.


2.3 - Média de avaliação dos clientes:

Embora a Loja 1 apresente a menor média de avaliação (3,98), a diferença em relação às demais é mínima. A média geral das quatro lojas é 4,02, com a maior nota sendo 4,05 – uma amplitude de apenas 0,07 pontos, essa variação é insignificante para impactar a decisão.


2.4 - Produtos mais e menos vendidos:

O dado mais relevante nessa análise é que a Loja 4 tem uma maior dependência das vendas de Camas Box, um produto que ocupa mais espaço no estoque, apresenta maior dificuldade logística (armazenamento e transporte). Podendo assim elevar os custos operacionais.

2.5 - Frete médio por Loja:

A loja 1 apresenta o frete médio mais caro R$34,69 podendo ter algum impacto sobre as vendas. Contudo a diferença em relação as outras lojas não é significativa, visto que a média de todas as lojas é de R33,16 e o valor minimo é de R31,28.
